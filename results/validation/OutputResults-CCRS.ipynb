```python
# this plots ccrs data so ignore it for gbov
def plotData(dataList,xvar,yvar,percentile,fileName) :
    fig = plt.figure()
    if yvar == 'sampleDASF' :
        plt.plot([0,6],[0,6],'k--')
        plt.ylim(0,1)
        plt.xlim(0,6)
        plt.fill_between(np.arange(0,6,0.1) ,np.minimum(0.75*np.arange(0,6,0.1),np.arange(0,6,0.1)-0.75),np.maximum(1.25*np.arange(0,6,0.1),np.arange(0,6,0.1)+0.75),color = 'gray',alpha = 0.1)
        plt.fill_between(np.arange(0,6,0.1) ,np.minimum(0.80*np.arange(0,6,0.1),np.arange(0,6,0.1)-0.50),np.maximum(1.20*np.arange(0,6,0.1),np.arange(0,6,0.1)+0.50),color = 'gray',alpha = 0.3)
        plt.fill_between(np.arange(0,6,0.1) ,np.minimum(0.85*np.arange(0,6,0.1),np.arange(0,6,0.1)-0.25),np.maximum(1.15*np.arange(0,6,0.1),np.arange(0,6,0.1)+0.25),color = 'gray',alpha = 0.5)
        plt.ylabel('DASF SL2P')
        plt.xlabel('LAI In-Situ')
    elif xvar == 'LAICEV61_Total' :
        plt.plot([0,6],[0,6],'k--')
        plt.ylim(0,6)
        plt.xlim(0,6)
        plt.fill_between(np.arange(0,6,0.1) ,np.minimum(0.75*np.arange(0,6,0.1),np.arange(0,6,0.1)-0.75),np.maximum(1.25*np.arange(0,6,0.1),np.arange(0,6,0.1)+0.75),color = 'gray',alpha = 0.1)
        plt.fill_between(np.arange(0,6,0.1) ,np.minimum(0.80*np.arange(0,6,0.1),np.arange(0,6,0.1)-0.50),np.maximum(1.20*np.arange(0,6,0.1),np.arange(0,6,0.1)+0.50),color = 'gray',alpha = 0.3)
        plt.fill_between(np.arange(0,6,0.1) ,np.minimum(0.85*np.arange(0,6,0.1),np.arange(0,6,0.1)-0.25),np.maximum(1.15*np.arange(0,6,0.1),np.arange(0,6,0.1)+0.25),color = 'gray',alpha = 0.5)
        plt.ylabel('LAI SL2P')
        plt.xlabel('LAI In-Situ')
    elif xvar == "fcover_Total" : 
        plt.plot([0,1],[0,1],'k--')
        plt.ylim(0,1)
        plt.xlim(0,1)
        plt.fill_between(np.arange(0,1,0.01) ,np.minimum(0.8*np.arange(0,1,0.01),np.arange(0,1,0.01)-0.10),np.maximum(1.2*np.arange(0,1,0.01),np.arange(0,1,0.01)+0.10),color = 'gray',alpha = 0.1)
        plt.fill_between(np.arange(0,1,0.01) ,np.minimum(0.85*np.arange(0,1,0.01),np.arange(0,1,0.01)-0.075),np.maximum(1.15*np.arange(0,1,0.01),np.arange(0,1,0.01)+0.075),color = 'gray',alpha = 0.3)
        plt.fill_between(np.arange(0,1,0.01) ,np.minimum(0.90*np.arange(0,1,0.01),np.arange(0,1,0.01)-0.05),np.maximum(1.10*np.arange(0,1,0.01),np.arange(0,1,0.01)+0.05),color = 'gray',alpha = 0.5)
        plt.ylabel('FCOVER SL2P')
        plt.xlabel('FCOVER In-Situ')
    elif xvar == 'Daily_Integrated_Black_Sky_fAPAR_Total' :
        plt.plot([0,1],[0,1],'k--')
        plt.ylim(0,1)
        plt.xlim(0,1)
        plt.fill_between(np.arange(0,1,0.01) ,np.minimum(0.8*np.arange(0,1,0.01),np.arange(0,1,0.01)-0.10),np.maximum(1.2*np.arange(0,1,0.01),np.arange(0,1,0.01)+0.10),color = 'gray',alpha = 0.1)
        plt.fill_between(np.arange(0,1,0.01) ,np.minimum(0.85*np.arange(0,1,0.01),np.arange(0,1,0.01)-0.075),np.maximum(1.15*np.arange(0,1,0.01),np.arange(0,1,0.01)+0.075),color = 'gray',alpha = 0.3)
        plt.fill_between(np.arange(0,1,0.01) ,np.minimum(0.95*np.arange(0,1,0.01),np.arange(0,1,0.01)-0.05),np.maximum(1.05*np.arange(0,1,0.01),np.arange(0,1,0.01)+0.05),color = 'gray',alpha = 0.5)
        plt.ylabel('FAPAR SL2P')
        plt.xlabel('FIPAR In-Situ')
    elif  xvar == 'PAIeCEV61_Total' :
        plt.plot([0,4],[0,4],'k--')
        plt.ylim(0,4)
        plt.xlim(0,4)
        plt.fill_between(np.arange(0,6,0.1) ,np.minimum(0.75*np.arange(0,6,0.1),np.arange(0,6,0.1)-0.75),np.maximum(1.25*np.arange(0,6,0.1),np.arange(0,6,0.1)+0.75),color = 'gray',alpha = 0.1)
        plt.fill_between(np.arange(0,6,0.1) ,np.minimum(0.80*np.arange(0,6,0.1),np.arange(0,6,0.1)-0.50),np.maximum(1.20*np.arange(0,6,0.1),np.arange(0,6,0.1)+0.50),color = 'gray',alpha = 0.3)
        plt.fill_between(np.arange(0,6,0.1) ,np.minimum(0.85*np.arange(0,6,0.1),np.arange(0,6,0.1)-0.25),np.maximum(1.15*np.arange(0,6,0.1),np.arange(0,6,0.1)+0.25),color = 'gray',alpha = 0.5)
        plt.ylabel('PAIe SL2P')
        plt.xlabel('PAIe In-Situ')
    else :
        plt.plot([0,1],[0,1],'k--')
        plt.ylim(0,1)
        plt.xlim(0,1)
        
    error = np.zeros(len(dataList))
    correct = np.zeros(len(dataList)) 
    for n in range(0,len(dataList)) :

        if ( (dataList[n][yvar])[2]['data'] )  and ( dataList[n]['PlotID'] != 'TP_plot1_T1_25JUN2019') and ( dataList[n]['PlotID'] != 'PR_plot1_T1_12AUG2019') and ( dataList[n]['PlotID'] != 'TP_plot4_T1_27JUN2019')  and ( dataList[n]['PlotID'] != 'G_C8_L1_R3'): 
            
            # get estimate and actual
            est = np.array((dataList[n][yvar])[2]['data'])/1000
            stdError = np.array((dataList[n][yvar])[5]['data'])/1000
            actual = np.array(dataList[n][xvar])
           
            # apply priors for LAI
            if yvar == 'sampleLAI' :                       
                if dataList[n]['NF'] == 1:
                     actual =  np.array(dataList[n]['PAICEV61_Up'])*(1-0.16)+np.array(dataList[n]['PAICEV61_Down'])
                elif dataList[n]['BF'] == 1:
                    actual =  np.array(dataList[n]['PAICEV61_Up'])*(1-0.24)+np.array(dataList[n]['PAICEV61_Down'])
                else :
                     actual =  np.array(dataList[n]['PAICEV61_Up'])*(1-0.19)+np.array(dataList[n]['PAICEV61_Down'])
            elif yvar == 'samplefCOVER' :  
                # correct actual for green fcover and get DASF based estimate
                if dataList[n]['NF'] == 1:
                    actual =  np.array(dataList[n]['fcover_Up'])*(1-0.16)+(1-np.array(dataList[n]['fcover_Up']))*np.array(dataList[n]['fcover_Down'])
                elif dataList[n]['BF'] == 1:
                    actual =  np.array(dataList[n]['fcover_Up'])*(1-0.24)+(1-np.array(dataList[n]['fcover_Up']))*np.array(dataList[n]['fcover_Down'])
                else :
                    actual =  np.array(dataList[n]['fcover_Up'])*(1-0.18)+(1-np.array(dataList[n]['fcover_Up']))*np.array(dataList[n]['fcover_Down'])
            elif yvar == 'samplefAPAR' :      
                # correct actual for green fAPAR and get DASF based estimate
                # fix issue with non standard key names
                if 'Daily_Integrated_Black_Sk_yfAPAR_Up' in dataList[n].keys() :
                    upName = 'Daily_Integrated_Black_Sk_yfAPAR_Up'
                else :
                    upName = 'Daily_Integrated_Black_Sky_fAPAR_Up'
                if 'Daily_Integrated_Black_Sk_yfAPAR_Down' in dataList[n].keys() :
                    downName = 'Daily_Integrated_Black_Sk_yfAPAR_Down'
                else :
                    downName = 'Daily_Integrated_Black_Sky_fAPAR_Down'
                    
                if dataList[n]['NF'] == 1:
                    actual =  np.array(dataList[n][upName])*(1-0.16)+(1-np.array(dataList[n][upName]))*np.array(dataList[n][downName]) 
                        
                elif dataList[n]['BF'] == 1:
                     actual =  np.array(dataList[n][upName])*(1-0.24)+(1-np.array(dataList[n][upName]))*np.array(dataList[n][downName]) 
                
                else :
                    actual =  np.array(dataList[n][upName])*(1-0.18)+(1-np.array(dataList[n][upName]))*np.array(dataList[n][downName])                        
                    
            residual = est-actual
            residualAbs = np.absolute(residual)
            index = np.where(residualAbs<np.percentile(residualAbs,percentile))
            filteredresiduals = residual[index[0]]
            stdError = np.median(stdError[index[0]])
            #print(filteredresiduals)
            error[n] = np.median(filteredresiduals)
            estimate = actual + np.median(filteredresiduals)
            x = actual
            y = estimate
            if xvar == 'LAICEV61_Total' :
                laierror = (np.array(dataList[n]['PAIEMiller_Total']) -np.maximum(np.array(dataList[n]['PAIeLAI2000(3)_Total']) ,np.array(dataList[n]['PAIeLAI2000(4)_Total']))) * np.array(dataList[n]['LAICEV61_Total']) / (0.1+np.array(dataList[n]['PAIeCEV61_Total'])) 
                xerr = (np.array(dataList[n]['LAICEV61std_total'])**2 +  (laierror/2)**2 )**(0.5)
                correct[n] = np.absolute(error[n])<np.maximum(0.25*actual,0.75)
            elif xvar == "fcover_Total" : 
                xerr = np.array(dataList[n]['fcoverstd_Total'])
                correct[n] = np.absolute(error[n])<np.maximum(0.20*actual,0.10)
            elif xvar == 'Daily_Integrated_Black_Sky_fAPAR_Total' :
                xerr = np.array(dataList[n]['fcoverstd_Total']);
                correct[n] = np.absolute(error[n])<np.maximum(0.20*actual,0.10)
            elif xvar == 'PAIeCEV61_Total' :
                laierror = (np.array(dataList[n]['PAIEMiller_Total']) - np.maximum(np.array(dataList[n]['PAIeLAI2000(3)_Total']) ,np.array(dataList[n]['PAIeLAI2000(4)_Total']))) 
                xerr = laierror/2
                correct[n] = np.absolute(error[n])<np.maximum(0.25*actual,0.75)
            else :
                xerr = 0;
        
            yerr = ((np.ptp(residual)/4/(len(residual)**0.5))**2 +  stdError**2)**0.5 #/(len(filteredresiduals)**0.5)
            #print((np.ptp(residual)/4/(len(residual)**0.5)) ,stdError)
            # colour based on needleaf fraction
            if dataList[n]['NF'] == 1:
                symbolColour = 'b'
            elif dataList[n]['BF'] == 1:
                symbolColour = 'r'
            else :
                symbolColour = 'y'

            # set symbol size to clumping index
            symbolSize = 20* (1- (np.array(dataList[n]['PAIeCEV61_Total']) /  (0.0001+np.array(dataList[n]['PAICEV61_Total']))))
            
            if dataList[n]['Location'] == 'Geraldton2020' :
                plt.plot(x,y,symbolColour+'o',markersize =symbolSize )
            elif dataList[n]['Location'] == 'HayRiver2019' :
                plt.plot(x,y,symbolColour+'>',markersize =symbolSize )
            elif dataList[n]['Location'] == 'Labrador2019' :
                plt.plot(x,y,symbolColour+'*',markersize =symbolSize )
            elif dataList[n]['Location'] == 'Merbleue2019' :
                plt.plot(x,y,symbolColour+'<',markersize =symbolSize )
            elif dataList[n]['Location'] == 'MtPolley2019' :
                plt.plot(x,y,symbolColour+'p',markersize =symbolSize )
            elif dataList[n]['Location'] == 'PeaceRiver2019' :
                plt.plot(x,y,symbolColour+'s',markersize =symbolSize )
            elif dataList[n]['Location'] == 'TurkeyPoint2019' :
                plt.plot(x,y,symbolColour+'d',markersize =symbolSize )
            elif dataList[n]['Location'] == 'Vancouver2019' :
                plt.plot(x,y,symbolColour+'v',markersize =symbolSize )
            elif dataList[n]['Location'] == 'Yellowknife2019' :
                plt.plot(x,y,symbolColour+'^',markersize =symbolSize )
            else :
                plt.plot(x,y,symbolColour+'o',markersize =symbolSize )

                            
            # circle lichen plots
            #if dataList[n]['Lichen'] >0.5:
            #     plt.plot(x,y,'co',markerfacecolor='none')
            #print(x,y,xerr,yerr)        
            plt.errorbar(x,y,yerr=yerr,xerr=xerr,fmt = 'k-.',linewidth=0.2)
            #print(n,actual,estimate,symbolSize )

    textstr = '\n'.join((
        r'$N=%d$' % (len(dataList), ),
        r'$MAE=%.2f$' % (np.median(np.absolute(error)), ),
        r'$U=%.2f$' % (np.power(np.mean(np.power(error,2)),0.5), ),
        r'$A=%.2f$' % (np.mean(error), ),
        r'$P=%.2f$' % (np.power(np.mean(np.power(np.subtract(error,np.mean(error)),2)),0.5), ),
        r'$THRESH=%.2f$%%' % (np.mean(correct)*100,)))
    ax = plt.gca()
    ax.text(0.015, 0.985, textstr,
        horizontalalignment='left',
        verticalalignment='top',
        transform=ax.transAxes)
    fig.savefig(fileName, bbox_inches='tight')
```

```python
# this plots ccrs data so ignore it for gbov
def plotDataD(datalistDASF,dataList,xvar,yvar,percentile,fileName) :
    fig = plt.figure()
    if yvar == 'sampleDASF' :
        plt.plot([0,6],[0,6],'k--')
        plt.ylim(0,1)
        plt.xlim(0,6)
        plt.fill_between(np.arange(0,6,0.1) ,np.minimum(0.75*np.arange(0,6,0.1),np.arange(0,6,0.1)-0.75),np.maximum(1.25*np.arange(0,6,0.1),np.arange(0,6,0.1)+0.75),color = 'gray',alpha = 0.1)
        plt.fill_between(np.arange(0,6,0.1) ,np.minimum(0.80*np.arange(0,6,0.1),np.arange(0,6,0.1)-0.50),np.maximum(1.20*np.arange(0,6,0.1),np.arange(0,6,0.1)+0.50),color = 'gray',alpha = 0.3)
        plt.fill_between(np.arange(0,6,0.1) ,np.minimum(0.85*np.arange(0,6,0.1),np.arange(0,6,0.1)-0.25),np.maximum(1.15*np.arange(0,6,0.1),np.arange(0,6,0.1)+0.25),color = 'gray',alpha = 0.5)
        plt.ylabel('DASF SL2P')
        plt.xlabel('LAI In-Situ')
    elif xvar == 'LAICEV61_Total' :
        plt.plot([0,6],[0,6],'k--')
        plt.ylim(0,6)
        plt.xlim(0,6)
        plt.fill_between(np.arange(0,6,0.1) ,np.minimum(0.75*np.arange(0,6,0.1),np.arange(0,6,0.1)-0.75),np.maximum(1.25*np.arange(0,6,0.1),np.arange(0,6,0.1)+0.75),color = 'gray',alpha = 0.1)
        plt.fill_between(np.arange(0,6,0.1) ,np.minimum(0.80*np.arange(0,6,0.1),np.arange(0,6,0.1)-0.50),np.maximum(1.20*np.arange(0,6,0.1),np.arange(0,6,0.1)+0.50),color = 'gray',alpha = 0.3)
        plt.fill_between(np.arange(0,6,0.1) ,np.minimum(0.85*np.arange(0,6,0.1),np.arange(0,6,0.1)-0.25),np.maximum(1.15*np.arange(0,6,0.1),np.arange(0,6,0.1)+0.25),color = 'gray',alpha = 0.5)
        plt.ylabel('LAI SL2P')
        plt.xlabel('LAI In-Situ')
    elif xvar == "fcover_Total" : 
        plt.plot([0,1],[0,1],'k--')
        plt.ylim(0,1)
        plt.xlim(0,1)
        plt.fill_between(np.arange(0,1,0.01) ,np.minimum(0.8*np.arange(0,1,0.01),np.arange(0,1,0.01)-0.10),np.maximum(1.2*np.arange(0,1,0.01),np.arange(0,1,0.01)+0.10),color = 'gray',alpha = 0.1)
        plt.fill_between(np.arange(0,1,0.01) ,np.minimum(0.85*np.arange(0,1,0.01),np.arange(0,1,0.01)-0.075),np.maximum(1.15*np.arange(0,1,0.01),np.arange(0,1,0.01)+0.075),color = 'gray',alpha = 0.3)
        plt.fill_between(np.arange(0,1,0.01) ,np.minimum(0.90*np.arange(0,1,0.01),np.arange(0,1,0.01)-0.05),np.maximum(1.10*np.arange(0,1,0.01),np.arange(0,1,0.01)+0.05),color = 'gray',alpha = 0.5)
        plt.ylabel('FCOVER SL2P')
        plt.xlabel('FCOVER In-Situ')
    elif xvar == 'Daily_Integrated_Black_Sky_fAPAR_Total' :
        plt.plot([0,1],[0,1],'k--')
        plt.ylim(0,1)
        plt.xlim(0,1)
        plt.fill_between(np.arange(0,1,0.01) ,np.minimum(0.8*np.arange(0,1,0.01),np.arange(0,1,0.01)-0.10),np.maximum(1.2*np.arange(0,1,0.01),np.arange(0,1,0.01)+0.10),color = 'gray',alpha = 0.1)
        plt.fill_between(np.arange(0,1,0.01) ,np.minimum(0.85*np.arange(0,1,0.01),np.arange(0,1,0.01)-0.075),np.maximum(1.15*np.arange(0,1,0.01),np.arange(0,1,0.01)+0.075),color = 'gray',alpha = 0.3)
        plt.fill_between(np.arange(0,1,0.01) ,np.minimum(0.95*np.arange(0,1,0.01),np.arange(0,1,0.01)-0.05),np.maximum(1.05*np.arange(0,1,0.01),np.arange(0,1,0.01)+0.05),color = 'gray',alpha = 0.5)
        plt.ylabel('FAPAR SL2P')
        plt.xlabel('FIPAR In-Situ')
    elif  xvar == 'PAIeCEV61_Total' :
        plt.plot([0,4],[0,4],'k--')
        plt.ylim(0,4)
        plt.xlim(0,4)
        plt.fill_between(np.arange(0,6,0.1) ,np.minimum(0.75*np.arange(0,6,0.1),np.arange(0,6,0.1)-0.75),np.maximum(1.25*np.arange(0,6,0.1),np.arange(0,6,0.1)+0.75),color = 'gray',alpha = 0.1)
        plt.fill_between(np.arange(0,6,0.1) ,np.minimum(0.80*np.arange(0,6,0.1),np.arange(0,6,0.1)-0.50),np.maximum(1.20*np.arange(0,6,0.1),np.arange(0,6,0.1)+0.50),color = 'gray',alpha = 0.3)
        plt.fill_between(np.arange(0,6,0.1) ,np.minimum(0.85*np.arange(0,6,0.1),np.arange(0,6,0.1)-0.25),np.maximum(1.15*np.arange(0,6,0.1),np.arange(0,6,0.1)+0.25),color = 'gray',alpha = 0.5)
        plt.ylabel('PAIe SL2P')
        plt.xlabel('PAIe In-Situ')
    else :
        plt.plot([0,1],[0,1],'k--')
        plt.ylim(0,1)
        plt.xlim(0,1)
        
    error = np.zeros(len(dataList))
    correct = np.zeros(len(dataList)) 
    for n in range(0,len(dataList)) :
        if ( (dataList[n][yvar])[2]['data'] )  and ( dataList[n]['PlotID'] != 'TP_plot1_T1_25JUN2019') and ( dataList[n]['PlotID'] != 'PR_plot1_T1_12AUG2019') and ( dataList[n]['PlotID'] != 'TP_plot4_T1_27JUN2019')  and ( dataList[n]['PlotID'] != 'G_C8_L1_R3'): 

            
            # get estimate and actual
            est = np.array((dataList[n][yvar])[2]['data'])/1000
            stdError = np.array((dataList[n][yvar])[5]['data'])/1000
            actual = np.array(dataList[n][xvar])
           
            # apply priors for LAI
            dasf = np.array((datalistDASF[n]['sampleDASF'])[2]['data'])/1000
            if yvar == 'sampleLAI' :                       
                if dataList[n]['NF'] == 1:
                    estDASF = np.exp(0.023505*(np.power(np.log(dasf),2)) + 1.7414*np.log(dasf) + 2.7345)
                    stdErrorDASF = 14.289*np.power(dasf,2)+4.2901*dasf-0.19569
                    actual =  np.array(dataList[n]['PAICEV61_Up'])*(1-0.16)+np.array(dataList[n]['PAICEV61_Down'])

                elif dataList[n]['BF'] == 1:
                    estDASF = np.exp(2.443e-5*(np.power(np.log(dasf),2)) + 1.578*np.log(dasf) + 2.383)
                    stdErrorDASF = 7.4952*np.power(dasf,2)+4.4447*dasf-0.22168
                    actual =  np.array(dataList[n]['PAICEV61_Up'])*(1-0.24)+np.array(dataList[n]['PAICEV61_Down'])

                else :
                    estDASF = np.exp(-0.0026808*(np.power(np.log(dasf),2)) + 1.4776*np.log(dasf) + 2.2686)
                    stdErrorDASF = 4.7536*np.power(dasf,2)+4.2083*dasf-0.18081  
                    actual =  np.array(dataList[n]['PAICEV61_Up'])*(1-0.19)+np.array(dataList[n]['PAICEV61_Down'])

                m1 = est
                v1 = np.power(stdError,2)
                m2 = estDASF
                v2 = np.power(stdErrorDASF,2)/4
                est = (m1*v2+m2*v1)/(v1+v2)
                stdError = (v1*v2)/(v1+v2)
            elif yvar == 'samplefCOVER' :  
                # correct actual for green fcover and get DASF based estimate
                if dataList[n]['NF'] == 1:
                    estDASF = (0.013748*(np.power(np.log(dasf),2)) + 0.27993*np.log(dasf) + 0.74465)
                    stdErrorDASF = -0.12315*np.power(dasf,2)+0.25902*dasf+0.12109
                    actual =  np.array(dataList[n]['fcover_Up'])*(1-0.16)+(1-np.array(dataList[n]['fcover_Up']))*np.array(dataList[n]['fcover_Down'])
                elif dataList[n]['BF'] == 1:
                    estDASF = (0.030859*(np.power(np.log(dasf),2)) + 0.3675*np.log(dasf) + 0.84891)
                    stdErrorDASF = -0.12347*np.power(dasf,2)+0.30826*dasf+0.131
                    actual =  np.array(dataList[n]['fcover_Up'])*(1-0.24)+(1-np.array(dataList[n]['fcover_Up']))*np.array(dataList[n]['fcover_Down'])
                else :
                    estDASF = (0.044048*(np.power(np.log(dasf),2)) + 0.41993*np.log(dasf) + 0.88904)
                    stdErrorDASF = -0.09133*np.power(dasf,2)+0.2678*dasf+0.11571  
                    actual =  np.array(dataList[n]['fcover_Up'])*(1-0.18)+(1-np.array(dataList[n]['fcover_Up']))*np.array(dataList[n]['fcover_Down'])
                m1 = est
                v1 = np.power(stdError,2)
                m2 = estDASF
                v2 = np.power(stdErrorDASF,2)/10
                est = (m1*v2+m2*v1)/(v1+v2)
                stdError = (v1*v2)/(v1+v2)
            elif yvar == 'samplefAPAR' :      
                # correct actual for green fAPAR and get DASF based estimate
                # fix issue with non standard key names
                if 'Daily_Integrated_Black_Sk_yfAPAR_Up' in dataList[n].keys() :
                    upName = 'Daily_Integrated_Black_Sk_yfAPAR_Up'
                else :
                    upName = 'Daily_Integrated_Black_Sky_fAPAR_Up'
                if 'Daily_Integrated_Black_Sk_yfAPAR_Down' in dataList[n].keys() :
                    downName = 'Daily_Integrated_Black_Sk_yfAPAR_Down'
                else :
                    downName = 'Daily_Integrated_Black_Sky_fAPAR_Down'
                    
                if dataList[n]['NF'] == 1:
                    estDASF = (0.069054*(np.power(np.log(dasf),2)) + 0.45712*np.log(dasf) + 0.77098)
                    stdErrorDASF = -0.025023*np.power(dasf,2)+0.15883*dasf+0.10932
                    actual =  np.array(dataList[n][upName])*(1-0.16)+(1-np.array(dataList[n][upName]))*np.array(dataList[n][downName]) 
                        
                elif dataList[n]['BF'] == 1:
                    estDASF = (0.079907*(np.power(np.log(dasf),2)) + 0.51654*np.log(dasf) + 0.85729)
                    stdErrorDASF = -0.02129*np.power(dasf,2)+0.21367*dasf+0.13461
                    actual =  np.array(dataList[n][upName])*(1-0.24)+(1-np.array(dataList[n][upName]))*np.array(dataList[n][downName]) 
                
                else :
                    estDASF = (0.094769*(np.power(np.log(dasf),2)) + 0.57588*np.log(dasf) + 0.89836)
                    stdErrorDASF = 0.0044235*np.power(dasf,2)+0.20181*dasf+0.13367  
                    actual =  np.array(dataList[n][upName])*(1-0.18)+(1-np.array(dataList[n][upName]))*np.array(dataList[n][downName])                        
                m1 = est
                v1 = np.power(stdError,2)
                m2 = estDASF
                v2 = np.power(stdErrorDASF,2)/4
                est = (m1*v2+m2*v1)/(v1+v2)
                stdError = (v1*v2)/(v1+v2)
                
            residual = est-actual
            residualAbs = np.absolute(residual)
            index = np.where(residualAbs<np.percentile(residualAbs,percentile))
            filteredresiduals = residual[index[0]]
            stdError = np.median(stdError[index[0]])
            #print(filteredresiduals)
            error[n] = np.median(filteredresiduals)
            estimate = actual + np.median(filteredresiduals)
            x = actual
            y = estimate
            if xvar == 'LAICEV61_Total' :
                laierror = (np.array(dataList[n]['PAIEMiller_Total']) -np.maximum(np.array(dataList[n]['PAIeLAI2000(3)_Total']) ,np.array(dataList[n]['PAIeLAI2000(4)_Total']))) * np.array(dataList[n]['LAICEV61_Total']) / (0.1+np.array(dataList[n]['PAIeCEV61_Total'])) 
                xerr = (np.array(dataList[n]['LAICEV61std_total'])**2 +  (laierror/2)**2 )**(0.5)
                correct[n] = np.absolute(error[n])<np.maximum(0.25*actual,0.75)
            elif xvar == "fcover_Total" : 
                xerr = np.array(dataList[n]['fcoverstd_Total'])
                correct[n] = np.absolute(error[n])<np.maximum(0.20*actual,0.10)
            elif xvar == 'Daily_Integrated_Black_Sky_fAPAR_Total' :
                xerr = np.array(dataList[n]['fcoverstd_Total']);
                correct[n] = np.absolute(error[n])<np.maximum(0.20*actual,0.10)
            elif xvar == 'PAIeCEV61_Total' :
                laierror = (np.array(dataList[n]['PAIEMiller_Total']) - np.maximum(np.array(dataList[n]['PAIeLAI2000(3)_Total']) ,np.array(dataList[n]['PAIeLAI2000(4)_Total']))) 
                xerr = laierror/2
                correct[n] = np.absolute(error[n])<np.maximum(0.25*actual,0.75)
            else :
                xerr = 0;
        
            yerr = ((np.ptp(residual)/4/(len(residual)**0.5))**2 +  stdError**2)**0.5 #/(len(filteredresiduals)**0.5)
            #print((np.ptp(residual)/4/(len(residual)**0.5)) ,stdError)
            # colour based on needleaf fraction
            if dataList[n]['NF'] == 1:
                symbolColour = 'b'
            elif dataList[n]['BF'] == 1:
                symbolColour = 'r'
            else :
                symbolColour = 'y'

            # set symbol size to clumping index
            symbolSize = 20* (1- (np.array(dataList[n]['PAIeCEV61_Total']) /  (0.0001+np.array(dataList[n]['PAICEV61_Total']))))
            
            if dataList[n]['Location'] == 'Geraldton2020' :
                plt.plot(x,y,symbolColour+'o',markersize =symbolSize )
            elif dataList[n]['Location'] == 'HayRiver2019' :
                plt.plot(x,y,symbolColour+'>',markersize =symbolSize )
            elif dataList[n]['Location'] == 'Labrador2019' :
                plt.plot(x,y,symbolColour+'*',markersize =symbolSize )
            elif dataList[n]['Location'] == 'Merbleue2019' :
                plt.plot(x,y,symbolColour+'<',markersize =symbolSize )
            elif dataList[n]['Location'] == 'MtPolley2019' :
                plt.plot(x,y,symbolColour+'p',markersize =symbolSize )
            elif dataList[n]['Location'] == 'PeaceRiver2019' :
                plt.plot(x,y,symbolColour+'s',markersize =symbolSize )
            elif dataList[n]['Location'] == 'TurkeyPoint2019' :
                plt.plot(x,y,symbolColour+'d',markersize =symbolSize )
            elif dataList[n]['Location'] == 'Vancouver2019' :
                plt.plot(x,y,symbolColour+'v',markersize =symbolSize )
            elif dataList[n]['Location'] == 'Yellowknife2019' :
                plt.plot(x,y,symbolColour+'^',markersize =symbolSize )
            else :
                plt.plot(x,y,symbolColour+'o',markersize =symbolSize )

                            
            # circle lichen plots
            #if dataList[n]['Lichen'] >0.5:
            #     plt.plot(x,y,'co',markerfacecolor='none')
            #print(x,y,xerr,yerr)        
            plt.errorbar(x,y,yerr=yerr,xerr=xerr,fmt = 'k-.',linewidth=0.2)
            #print(n,actual,estimate,symbolSize )

    textstr = '\n'.join((
        r'$N=%d$' % (len(dataList), ),
        r'$MAE=%.2f$' % (np.median(np.absolute(error)), ),
        r'$U=%.2f$' % (np.power(np.mean(np.power(error,2)),0.5), ),
        r'$A=%.2f$' % (np.mean(error), ),
        r'$P=%.2f$' % (np.power(np.mean(np.power(np.subtract(error,np.mean(error)),2)),0.5), ),
        r'$THRESH=%.2f$%%' % (np.mean(correct)*100,)))
    ax = plt.gca()
    ax.text(0.015, 0.985, textstr,
        horizontalalignment='left',
        verticalalignment='top',
        transform=ax.transAxes)
    fig.savefig(fileName, bbox_inches='tight')
```
